### **Подключаем пакеты и вспомогательные файлы**

```julia
include("makePlot.jl")
using DelimitedFiles
folder = "D:\\Calc_Methods_2\\Lab3\\CrossScheme\\"
```

### **Выбираем номер уравнения**

```julia
numOfEq = 1
println("Номер уравнения: $numOfEq")
```

### **Загрузка результатов программы**

```julia
solutionFile = folder * "solution$numOfEq.txt"
dataFile = folder * "data$numOfEq.txt"
intervalFile = folder * "interval$numOfEq.txt"

solutionMatrix = readdlm(solutionFile)
dataVec = readdlm(dataFile)
intervalVec = readdlm(intervalFile);

a = dataVec[1]
L = dataVec[2]
timeEnd = dataVec[3]
numOfXIntervals = Int(intervalVec[1])
numOfTIntervals = Int(intervalVec[2])
h = L / numOfXIntervals
τ = timeEnd / numOfTIntervals

print("a = $a", '\n')
print("L = $L", '\n')
print("timeEnd = $timeEnd", '\n')
print("numOfXIntervals = $numOfXIntervals", '\n')
print("numOfTIntervals = $numOfTIntervals", '\n')
print("h = $h", '\n')
print("τ = $τ", '\n')
```

### **Параметры вывода графика**

```julia
numOfPlots = 100 # Сколько показать первых графиков // Работает только при default = false
plotStep = 10 # Шаг между первыми графиками // Работает только при default = false
println("Шаг графика: $plotStep")
println("Время последнего графика: $((numOfPlots - 1) * τ)")
```

### **Вывод графика**

```julia
if  numOfPlots <= numOfTIntervals
    makePlot(numOfPlots, plotStep, solutionMatrix, dataVec, intervalVec) 
else
    print("Количество графиков превышает количество временных слоев")
end
```
